```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
data = np.random.standard_normal(10000)
```

```python
n, bins, patches = plt.hist(data, 30, histtype='step', facecolor='g', color='g', alpha=0.5, label='standard normal')
```

```python
n2, bins2, patches2 = plt.hist(data, bins=np.arange(-5., 5.25, 0.25), histtype='stepfilled', color='b', alpha=0.8)
patches2[0].axes.get_figure().set_dpi(200)
patches2[0].axes.get_figure().set_size_inches(8,6)
patches2[0].axes.set_xlabel('x')
patches2[0].axes.set_ylabel('counts / 0.25')
patches2[0].axes.set_title('standard normal distribution')
```
